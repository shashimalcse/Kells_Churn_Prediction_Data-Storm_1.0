```python
sales_summary['basket_value']=sales_summary['total_amount_spent']/sales_summary['total_num_of_visits']
sales_summary['basket_value_weekend']=sales_summary['weekend_amount_spent']/sales_summary['Num_of_weekend_visits']
```

```python
sales_summary['time']=sales_summary['year'].astype(str)+"-"+sales_summary['month'].astype(str)+"-"+"01"

sales_summary['time']=pd.to_datetime(sales_summary['time'])

basket_mean=sales_summary.groupby(['customer_id'])['basket_value'].mean()
basket_mean=basket_mean.to_frame()
```

```python
rows,columns = df2.shape
print (columns)
df2.head()
df2.iloc[0][11]
```

```python
# df3 = df2.head()
# column = list(df3)
# print(df3.iloc[0,2])
# for i in range(5):
#     for j in range(35):
#         print(df3.iloc[i,j])
flagList = []
for i in range(rows):
    count = 0
    for j in range(columns-2):
        if df2.iloc[i][j] == 0:
            continue
        elif ((df2.iloc[i][j+1]==0) and (df2.iloc[i][j+2]==0)):
            count += 1
    flagList.append(count)
# print (flagList)
            
```

```python
a_series = pd.Series(flagList)
df2['flag'] = a_series.values
```

```python
df2.head()
```

```python
tot_frq = []
frq1 = []
frq2=[]
frq3=[]
for i in range(rows):
    temp_count = 0
    temp_count1 = 0
    temp_count2 = 0
    temp_count3 = 0
    for j in range(0,12):
        if df2.iloc[i][j]!=0:
            temp = 1
            temp_count += 1
            temp_count1 += 1
        else:
            continue
                
    for k in range(12,24):
        if df2.iloc[i][k]!=0:
            temp = 1
            temp_count += 1
            temp_count2 += 1
        else:
            continue
    
    for l in range(24,36):
        if df2.iloc[i][l]!=0:
            temp = 1
            temp_count += 1
            temp_count3 += 1
        else:
            continue
                
    tot_frq.append(temp_count/12)
    frq1.append(temp_count1/12)
    frq2.append(temp_count2/12)
    frq3.append(temp_count3/12)
```

```python
fr = pd.Series(tot_frq)
df2['frequency'] = fr.values

fr1 = pd.Series(frq1)
df2['2017 frequency'] = fr1.values

fr2 = pd.Series(frq2)
df2['2018 frequency'] = fr2.values

fr3 = pd.Series(frq3)
df2['2019 frequency'] = fr3.values
df2.head()
```

```python
valList = []
for i in range(rows):
    if (df2.iloc[i][12] == 0 and df2.iloc[i][13] == 0):
        val = 1
    else:
        val = 0
    valList.append(val)
va = pd.Series(valList)
df2['train1'] = va.values
df2.head()

valList1 = []
for i in range(rows):
    if (df2.iloc[i][24] == 0 and df2.iloc[i][25] == 0):
        val1 = 1
    else:
        val1 = 0
    valList1.append(val1)
va1 = pd.Series(valList1)
df2['train2'] = va1.values
df2.head()
```
